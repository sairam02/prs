```python
## python code to show boxcox transformation of non normal data

#import modules
import numpy as np
from scipy import stats

#plotting modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## generate non-normal data (exponential)
original_data = np.random.exponential(size = 1000)

## transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(original_data)


#plotting the original data(non-normal) and fitted data (normal)
sns.displot(original_data, kde = True,
           rug_kws = {'shade':True, 'linewidth':2},
           label = 'Non-Normal', color = "green")

sns.displot(fitted_data, kde = True,
           rug_kws = {'shade':True, 'linewidth':2},
           label = 'Normal', color = "green")

#adding legends to the subplots
plt.legend(loc = "upper right")


print(f"Lambda value used for Transformation: {fitted_lambda}")
```

```python
sm.qqplot(fitted_data, line = ('45'))
py.show()
```
